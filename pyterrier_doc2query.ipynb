```python
import pyterrier as pt ; pt.init()
import pandas as pd
from pyterrier_doc2query import Doc2Query, QueryScorer, QueryFilter
from pyterrier_dr import ElectraScorer
```

```python
doc2query = Doc2Query('macavaney/doc2query-t5-base-msmarco', append=False, num_samples=5)
scorer = ElectraScorer('crystina-z/monoELECTRA_LCE_nneg31')
pipeline = doc2query >> QueryScorer(scorer) >> QueryFilter(append=False, t=-2.21679688)
# use append=True when indexing; t=-2.21679688 is the 10th percentile for generated queries on MS MARCO
```
